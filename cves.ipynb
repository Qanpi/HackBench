```python
import os
import pandas as pd
import json
```

```python
stop = False

rows = []

for root, dirs, files in os.walk('cves/'):
    # FIXME: only 2024 for now
    if '2024' not in root:
        continue

    for file in files:
        if file.startswith('CVE') and file.endswith('.json'):
            file_path = os.path.join(root, file)

            with open(file_path, 'r') as f:
                parsed = json.load(f)
                rows.append(parsed)

df = pd.json_normalize(rows, sep='_')
```

```python
# Create a new dataframe with better aligned columns
cve_df = pd.DataFrame({
    'cve_id': df['cveMetadata_cveId'],
    'title': df['containers_cna_title'],
    'description': df['containers_cna_descriptions'].apply(lambda x: x['value']),
    'severity': df['containers_cna_metrics'].apply(lambda x: x[0]['cvssV3_1']['baseSeverity'] if x and len(x) > 0 and 'cvssV3_1' in x[0] else None),
    'score': df['containers_cna_metrics'].apply(lambda x: x[0]['cvssV3_1']['baseScore'] if x and len(x) > 0 and 'cvssV3_1' in x[0] else None),
    'date_published': pd.to_datetime(df['cveMetadata_datePublished']),
    'problem_type': df['containers_cna_problemTypes'].apply(lambda x: x[0]['descriptions'][0]['description'] if x and len(x) > 0 and 'descriptions' in x[0] else None),
    'vendor': df['containers_cna_affected'].apply(lambda x: x[0]['vendor'] if x and len(x) > 0 else None),
    'product': df['containers_cna_affected'].apply(lambda x: x[0]['product'] if x and len(x) > 0 else None),
    'solution': df['containers_cna_solutions'].apply(lambda x: x[0]['value'] if x and len(x) > 0 else None)
})

# Display the first few rows to verify
cve_df.head()
```
